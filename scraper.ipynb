# This module scrapes the script to find conversations with Michael
### Source: officequotes.net

```python
from urllib3 import PoolManager
from bs4 import BeautifulSoup
import re
```

```python
http = urllib3.PoolManager()
url = 'http://officequotes.net/no1-01.php'
```

```python
page = http.request('GET',url)
```

```python
soup = BeautifulSoup(page.data, 'html.parser')
```

```python
print(soup.prettify())
```

```python
quotes = soup.findAll("div", attrs = {'class':'quote'})
```

To clean this data, we have to remove:
1. HTML Tags
2. Actions mentioned in [..]

```python
quote = quotes[0].text.strip()
```

```python
out = re.findall(r'(\[.*?\])', quote)
```

```python
for o in out:
    quote = quote.replace(o, "")
print (quote)
```

```python
names = []
dialogues = quote.split('\n')
for idx,dialog in enumerate(dialogues):
    name = re.findall(r'(^\w+: )', dialog)
    dialogues[idx] = dialog.replace(name[0], "")
    names.append(name[0][:-2])
print(names)
print (dialogues)
```

```python
data = []
for idx,name in enumerate(names):
    if idx > 0 and name == "Michael":
        x = dialogues[idx - 1]
        y = dialogues[idx]
        data.append((x,y))
print (data)
```
